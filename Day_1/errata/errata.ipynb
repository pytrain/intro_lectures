# What's with the . (DOT)?

---

The . (dot) ties together the functionality of whatever preceeds it. Here's an example:

```python
a_string = 'some test string'
```

```python
a_string.capitalize()
```

```python
a_dict = {'a':'dictionary'}
```

```python
a_dict.capitalize()
```

This is due to Object-Oriented Programming (OOP) that specifies the functionality or methods associated with a particular object (or data type). To see what functionality is available for a particular variable, expression, etc.

```python
print(dir(a_string))
```

```python
print(dir(a_dict))
```
